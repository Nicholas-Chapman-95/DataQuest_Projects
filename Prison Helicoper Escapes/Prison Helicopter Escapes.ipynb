## Get the Data

```python
url = 'https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes'
data = data_from_url(url)
```

Let's print the first three rows

```python
for row in data[:3]:
    print(row)
```

```python
index = 0

for row in data:
    data[index] = row[:-1] ## Removes the last column
    index += 1 
```

```python
for row in data:
    row[0] = fetch_year(row[0]) ## Why does this work when using a second for loop?
```

```python
print(data[:2]) ## First 2 rows of data
```

```python
min_year = min(data, key=lambda x: x[0])[0]
max_year = max(data, key=lambda x: x[0])[0]
```

```python
years = []
for year in range(min_year, max_year + 1):
    years.append(year)
```

```python
##print(years)
print(years[:2])

attempts_per_year = []
for year in years:
    attempts_per_year.append([year, 0])


print(attempts_per_year)

## length of numbers where year = data.year?
## count where equal?

##attempts_per_year = []
```

```python
for row in data:
    for year_attempt in attempts_per_year:
        year = year_attempt[0]
        if row[0] == year:
            year_attempt[1] += 1
            
            
print(attempts_per_year)

```
